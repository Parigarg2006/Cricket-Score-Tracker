Cricket Score Tracker

```python
from dataclasses import dataclass, field
from typing import List
import random

@dataclass
class Batsman:
    name: str
    runs: int = 0
    balls: int = 0
    fours: int = 0
    sixes: int = 0

    @property
    def strike_rate(self):
        return (self.runs / self.balls * 100) if self.balls else 0

    def face_ball(self, runs):
        self.balls += 1
        self.runs += runs
        if runs == 4: self.fours += 1
        if runs == 6: self.sixes += 1

@dataclass
class Innings:
    team: str
    overs: int = 20
    runs: int = 0
    wickets: int = 0
    balls: int = 0
    scorecard: List[Batsman] = field(default_factory=list)

    def add_ball(self, batsman, runs, wicket=False):
        self.balls += 1
        self.runs += runs
        if wicket: self.wickets += 1
        batsman.face_ball(runs)

    @property
    def current_over(self): return f"{self.balls // 6}.{self.balls % 6}"

    @property
    def run_rate(self): return (self.runs / self.balls * 6) if self.balls else 0

def simulate_innings(team_name, target=None):
    inn = Innings(team_name)
    players = [Batsman(f"Player {i+1}") for i in range(11)]
    current, partner = 0, 1
    inn.scorecard = players

    random.seed(42 if team_name == "India" else 7)
    for ball in range(120):
        if inn.wickets >= 10: break
        outcomes = [0,0,0,1,1,1,1,2,2,3,4,4,4,6,6,'W']
        result = random.choice(outcomes)
        if result == 'W':
            inn.add_ball(players[current], 0, wicket=True)
            current = partner + 1
            partner = current + 1
        else:
            inn.add_ball(players[current], result)
            if result % 2 == 1: current, partner = partner, current

    print(f"\n{'='*50}")
    print(f"  {team_name}: {inn.runs}/{inn.wickets} ({inn.current_over} overs)")
    print(f"  Run Rate: {inn.run_rate:.2f}")
    print(f"{'='*50}")
    print(f"{'Batsman':<15} {'R':>5} {'B':>5} {'4s':>4} {'6s':>4} {'SR':>7}")
    print("-" * 40)
    for p in players[:inn.wickets+2]:
        print(f"{p.name:<15} {p.runs:>5} {p.balls:>5} {p.fours:>4} {p.sixes:>4} {p.strike_rate:>7.1f}")
    return inn

inn1 = simulate_innings("India")
inn2 = simulate_innings("Australia", target=inn1.runs+1)
if inn2.runs > inn1.runs:
    print(f"\n🏆 Australia wins by {10-inn2.wickets} wickets!")
elif inn2.runs < inn1.runs:
    print(f"\n🏆 India wins by {inn1.runs - inn2.runs} runs!")
else:
    print("\n🤝 Match Tied!")
```
